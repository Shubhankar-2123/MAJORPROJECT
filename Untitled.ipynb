```python
# ===============================
# 1️⃣ Import Libraries
# ===============================
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from skorch import NeuralNetClassifier
import torch
import torch.nn as nn
import torch.nn.functional as F
import joblib
import matplotlib.pyplot as plt

# ===============================
# 2️⃣ Load CSV
# ===============================
df = pd.read_csv("data_keypoints/all_keypoints.csv")
print("Total rows:", len(df))
print(df['type'].value_counts())

# ===============================
# 3️⃣ Filter static images only
# ===============================
df_images = df[df['type'] == 'image']
print("Static image rows:", len(df_images))

# ===============================
# 4️⃣ Remove classes with few samples (<10)
# ===============================
counts = df_images['label'].value_counts()
valid_labels = counts[counts >= 10].index
df_images = df_images[df_images['label'].isin(valid_labels)]
print("Remaining classes:", df_images['label'].nunique())

# ===============================
# 5️⃣ Prepare features and labels
# ===============================
X = df_images.drop(columns=['label','type','file','frame']).values
y = df_images['label'].values

# Encode labels
le = LabelEncoder()
y = le.fit_transform(y)
num_classes = len(np.unique(y))
print("Number of classes after filtering:", num_classes)

# ===============================
# 6️⃣ Split train/test
# ===============================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
print("X_train:", X_train.shape, "X_test:", X_test.shape)

# ===============================
# 7️⃣ Normalize features
# ===============================
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# ===============================
# 8️⃣ Define MLP model (PyTorch)
# ===============================
class StaticSignModel(nn.Module):
    def __init__(self, input_dim, num_classes):
        super().__init__()
        self.fc1 = nn.Linear(input_dim, 512)
        self.fc2 = nn.Linear(512, 256)
        self.fc3 = nn.Linear(256, 128)
        self.fc4 = nn.Linear(128, num_classes)
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = F.relu(self.fc2(x))
        x = self.dropout(x)
        x = F.relu(self.fc3(x))
        x = self.dropout(x)
        return self.fc4(x)  # logits

# ===============================
# 9️⃣ Wrap in Skorch
# ===============================
input_dim = X_train.shape[1]

net = NeuralNetClassifier(
    StaticSignModel,
    module__input_dim=input_dim,
    module__num_classes=num_classes,
    max_epochs=50,
    lr=0.0005,
    batch_size=32,
    optimizer=torch.optim.AdamW,
    iterator_train__shuffle=True,
    verbose=1,
    device='cuda' if torch.cuda.is_available() else 'cpu'
)

# ===============================
# 10️⃣ Train Model
# ===============================
net.fit(X_train.astype(np.float32), y_train.astype(np.longlong))

# ===============================
# 11️⃣ Evaluate
# ===============================
y_pred = net.predict(X_test.astype(np.float32))
print("🎯 Test Accuracy:", accuracy_score(y_test, y_pred))
print("\n📊 Classification Report:\n", classification_report(y_test, y_pred, zero_division=0))

# ===============================
# 12️⃣ Save Model, Scaler, LabelEncoder
# ===============================
joblib.dump(net, "models/static_sign_model.pkl")
joblib.dump(scaler, "models/static_scaler.pkl")
joblib.dump(le, "models/static_label_encoder.pkl")
print("💾 Model, scaler, and label encoder saved successfully")
```
